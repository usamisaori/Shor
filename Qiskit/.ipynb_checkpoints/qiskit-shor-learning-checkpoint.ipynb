```python
vqcCircuit = QuantumCircuit(qubits_num)

index = 0
for layer in range(layers_num):
    for i in range(qubits_num):
        vqcCircuit.u2(params[index], params[index + 1], i)
        index += 2

    if layer + 1 == layers_num:
        break

    for i in range(qubits_num):
        if i + 1 != qubits_num:
            vqcCircuit.cx(i, i + 1)
        elif qubits_num > 2:
            vqcCircuit.cx(i, 0)
    vqcCircuit.barrier()

vqcCircuit.draw(output='mpl')
```

## Input Circuit

```python
def generateInputCircuit(num):
    inputCircuit = QuantumCircuit(qubits_num, qubits_num)
    
    numStr = bin(num)[2:]
    
    for (i, ch) in enumerate(numStr):
        if ch == '1':
            inputCircuit.x(i)
    
    return inputCircuit
```

```python
inputCircuit_1011 = generateInputCircuit(11)
inputCircuit_1011.draw(output='mpl')
```

## Measurements

```python
from functools import reduce

Dag = lambda matrix: matrix.conj().T
Kron = lambda *matrices: reduce(np.kron, matrices)
```

```python
def getMeasurements(n):
    psi_0 = np.array([1.0, 0.0])
    psi_1 = np.array([0.0, 1.0])
    I = np.eye(2)

    M_0 = psi_0.reshape([2, 1]) @ psi_0.reshape([1, 2]).conj()
    M_1 = psi_1.reshape([2, 1]) @ psi_1.reshape([1, 2]).conj()
    
    M = [M_0, M_1]
    
    measurements = []
    for i in range(2 ** n):
        binnum = bin(i)[2:].rjust(n, '0')

        temp = []
        indices = map(lambda x: int(x), list(binnum))
        for index in indices:
            temp.append(M[index])
        
        measurements.append( Kron(*temp) )
    
    return measurements
```

```python
measurements = getMeasurements(qubits_num)
print(len(measurements))
```

```python
from qiskit.quantum_info import DensityMatrix, Statevector

def getDensityMatrix(circuit):
    return DensityMatrix(circuit).data

def getStatevector(circuit):
    return Statevector(circuit).data
```

```python
def measure(M, state):
    return Dag(state) @ Dag(M) @ M @ state
```

```python
class VQCLearner:
    """
    circuits: [inputCircuit, *, outputCircuit]
    n: dimension about target operator, pos: eg. [0, 1, 2], len(pos) should equal to n
    expected: single num => target , list [...] => probability distribution
    """
    def __init__(self, circuits, n, expected):
        self.circuits = circuits
        self.n = n
        self.measurements = getMeasurements(n
        self.pos = range(n)
        self.expected = expected # functionMap
        
        self.params = []
        self.losses = []
        
        self.vqc_layers_num = 3
        
    def cost(self, num, circuit):
        target = self.expected[num]
        expecetd = int(
        ''.join(
            list(reversed(
                list(bin(num)[2:].zfill(phaseBitsQubits) + 
                     bin(target)[2:].zfill(functionValueQubits))
            ))
        ), 2)
        
        p = 
    
        
    def L(self, p):
        loss = 0.0
        
        # input circuit
        for (num, circuit) in enumerate(self.circuits):
            index = 0
            for layer in range(self.vqc_layers_num):
                for i in range(self.n):
                    circuit.u2(p[index], p[index + 1], self.pos[i])
                    index += 2

                if layer + 1 == self.vqc_layers_num:
                    break

                for i in range(self.n):
                    if i + 1 != self.n:
                        circuit.cx(self.pos[i], self.pos[i + 1])
                    elif self.n > 2:
                        circuit.cx(self.pos[i], self.pos[0])

            loss += cost(num, circuit) 

            return loss
        
    def gradientCalculator(self, p, epsilon):
        derivates = np.zeros(len(p))
        
        l_p = self.L(p)
        for i, param in enumerate(p):
            p[i] += epsilon
            l_pe = self.L(p)
            p[i] -= epsilon

            derivates[i] = (l_pe - l_p)

        return derivates / epsilon, l_p
    
    def fit(self, *, epoch=200, epsilon=0.01, stepsize=0.01, strategy='unitary'):
        self.params = []
        self.losses = []
        
        # strategy: unitary => buildU / VQC
        self.strategy = strategy
        
        # 1. initialize the parameters with uniformly random nunbers
        size = self.vqc_layers_num * self.n * 2
        p = np.random.uniform(-1, 1, size)

        # Adam parameters:
        momentum, s = 0, 0
        ## Stepsize
        alpha = stepsize
        belta1 = 0.9; belta2 = 0.999
        e = 1e-8

        t = 0 # epoch counter
        
        lastloss = 0
        while t < epoch:
            # repeat
            t += 1
            grad, loss = self.gradientCalculator(p, epsilon)
            self.losses.append(loss)
            self.params.append(p)

            if abs(loss - lastloss) < 1e-5:
                break
            lastloss = loss

            # Adam Optimizer
            momentum = belta1 * momentum + (1 - belta1) * grad
            s = belta2 * s + (1 - belta2) * (grad ** 2)
            m_ = momentum / (1 - belta1 ** t)
            s_ = s / (1 - belta2 ** t)

            # update parameters
            p = p - alpha * (m_ / ( s_ ** 0.5 + e ))

            # log
            print('epoch: [{}/{}] - loss: {} end.'.format(t, epoch, loss))
```
